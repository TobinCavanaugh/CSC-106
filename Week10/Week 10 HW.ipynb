```python
raw_df.plot(kind="line")
```

```python
raw_df.plot(kind="bar")
```

```python
raw_df.plot.hist()
```

```python
raw_df.plot(kind="box")
```

```python
raw_df.plot(kind="area")
```

```python
raw_df.plot(kind="scatter", x=0, y=1)
```
